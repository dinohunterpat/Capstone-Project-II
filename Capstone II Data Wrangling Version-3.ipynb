```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
autism_data = pd.read_csv('Autism_Data.arff')
```

```python
autism_data.info()
```

```python
# Check out Country of Residence
autism_data['contry_of_res']
```

```python
# List Column Names
autism_data.columns
```

```python
# Change Column Titles
list_labels = ['A1 Score', 'A2 Score', 'A3 Score', 'A4 Score', 'A5 Score', 'A6 Score',
'A7 Score', 'A8 Score', 'A9 Score', 'A10 Score', 'Age', 'Gender',
'Ethnicity', 'Jaundice', 'Autism in Family', 'Country of Residence', 
'Used App Before','Result', 'Age Description', 'Relation to Patient', 'ASD']
autism_data.columns = list_labels
autism_data.head()
```

```python
# Remove commas
autism_data['Country of Residence'] = autism_data['Country of Residence'].str.replace(r"[\"\',]", '')
autism_data['Age Description'] = autism_data['Age Description'].str.replace(r"[\"\',]", '')
autism_data['Ethnicity'] = autism_data['Ethnicity'].str.replace(r"[\"\',]", '')
autism_data['Relation to Patient'] = autism_data['Relation to Patient'].str.replace(r"[\"\',]", '')
```

```python
# Convert Relevant Column Values to String
autism_data[['Ethnicity','Gender','Country of Residence','ASD']] = autism_data[['Ethnicity','Gender','Country of Residence','ASD']].astype(str)
```

```python
autism_data.head()
```

```python
# Get info
autism_data.info()
```

```python
# Replace Values
autism_data = autism_data.replace(to_replace ="?", value ="Not Available")
autism_data = autism_data.replace(to_replace ="others", value ="Not Available")
autism_data.head()
```

```python
# Clean up Ethnicity column
autism_data['Ethnicity'] = autism_data['Ethnicity'].replace(to_replace ="others", value ="Other")
autism_data['Ethnicity'] = autism_data['Ethnicity'].replace(to_replace ="Others", value ="Other")
autism_data['Ethnicity'] = autism_data['Ethnicity'].replace(to_replace ="Pasifika", value ="Oceania")
#autism_data['Ethnicity'] = autism_data['Ethnicity'].replace(to_replace ="Turkish", value ="Other")
autism_data['Ethnicity'] = autism_data['Ethnicity'].replace(to_replace ="Latino", value ="Hispanic/Latino")
autism_data['Ethnicity'] = autism_data['Ethnicity'].replace(to_replace ="Hispanic", value ="Hispanic/Latino")
#autism_data['Ethnicity'] = autism_data['Ethnicity'].replace(to_replace ="Oceania", value ="Other")
```

```python
# Check Ethnicity Counts
autism_data.groupby(by=['Ethnicity'])['Ethnicity'].count()
```

```python
# Drop Small Values
autism_data.drop(autism_data.loc[autism_data['Ethnicity']=='Oceania'].index, inplace=True)
autism_data.drop(autism_data.loc[autism_data['Ethnicity']=='Turkish'].index, inplace=True)
```

```python
# Clean up Gender column
autism_data['Gender'] = autism_data['Gender'].str.capitalize()
autism_data['Gender'] = autism_data['Gender'].str.capitalize()
```

```python
# Convert Main Set to Dataframes
autsim_data = pd.DataFrame(autism_data)

# Yes and No Dataframes for ASD
autism_yes = autism_data[autism_data['Result'] > 6]  
autism_no = autism_data[(autism_data['Result'] <= 6)] 
```

```python
# Dataframe for Nationality of People with Autism Yes
ASD_country_percent = (autism_yes['Country of Residence'].value_counts()/autism_data['Country of Residence'].value_counts())*100
ASD_country_percent = ASD_country_percent.dropna()
ASD_country_percent_df = pd.DataFrame(ASD_country_percent)
country_score = pd.DataFrame(autism_yes.groupby(by=['Country of Residence'])['Result'].mean())
country_count = pd.DataFrame(autism_yes.groupby(by=['Country of Residence'])['Country of Residence'].count())
ASD_Yes_Country_df = pd.concat([ASD_country_percent_df,country_count,country_score], axis=1, ignore_index = True)
ASD_Yes_Country_df = ASD_Yes_Country_df.reset_index()
ASD_Yes_Country_df.columns = ['Country','Percentage with Autism','Number of People with Autism','Mean Autism Score']
ASD_Yes_Country_df.head()
```

```python
autism_data.groupby(by=['Ethnicity', 'ASD'])['ASD'].count()
```

```python
# Dataframe based on Ethnicity
ASD_ethnicity_yes_percent = (autism_yes['Ethnicity'].value_counts()/autism_data['Ethnicity'].value_counts())*100
ASD_ethnicity_no_percent = (autism_no['Ethnicity'].value_counts()/autism_data['Ethnicity'].value_counts())*100
ASD_ethnicity_yes_percent = ASD_ethnicity_yes_percent.dropna()
ASD_ethnicity_no_percent = ASD_ethnicity_no_percent.dropna()
ASD_ethnicity_all_score = autism_data.groupby(by=['Ethnicity'])['Result'].mean()
ASD_ethnicity_yes_score = autism_yes.groupby(by=['Ethnicity'])['Result'].mean()
ASD_ethnicity_no_score = autism_no.groupby(by=['Ethnicity'])['Result'].mean()
ASD_ethnicity_yes_count = autism_yes.groupby(by=['Ethnicity'])['Ethnicity'].count()
ASD_ethnicity_no_count = autism_no.groupby(by=['Ethnicity'])['Ethnicity'].count()
ASD_ethnicity_all_count = autism_data.groupby(by=['Ethnicity'])['Ethnicity'].count()

ASD_ethnicity_yes_percent = pd.DataFrame(ASD_ethnicity_yes_percent)
ASD_ethnicity_no_percent = pd.DataFrame(ASD_ethnicity_no_percent)

ASD_Ethnicity_df = pd.concat([ASD_ethnicity_yes_percent,
                              ASD_ethnicity_no_percent,
                              ASD_ethnicity_yes_count,
                              ASD_ethnicity_no_count,
                              ASD_ethnicity_all_count,
                              ASD_ethnicity_yes_score,
                              ASD_ethnicity_no_score,
                              ASD_ethnicity_all_score
                             ], axis=1, ignore_index = True)

ASD_Ethnicity_df = ASD_Ethnicity_df.reset_index()
ASD_Ethnicity_df.columns = ['Ethnicity','Percentage with Autism',
                            'Percentage without Autism',
                            'Number of People with Autism',
                            'Number of People without Autism',
                            'Total Number of People',
                            'Mean Autism Score for People with Autism',
                            'Mean Autism Score for People without Autism',
                            'Mean Autism Score for All People']
ASD_Ethnicity_df.head()
```

```python
# Dataframe based on Gender
ASD_gender_yes_percent = (autism_yes['Gender'].value_counts()/autism_data['Gender'].value_counts())*100
ASD_gender_no_percent = (autism_no['Gender'].value_counts()/autism_data['Gender'].value_counts())*100
ASD_gender_yes_percent = ASD_gender_yes_percent.dropna()
ASD_gender_no_percent = ASD_gender_no_percent.dropna()
ASD_gender_all_score = autism_data.groupby(by=['Gender'])['Result'].mean()
ASD_gender_yes_score = autism_yes.groupby(by=['Gender'])['Result'].mean()
ASD_gender_no_score = autism_no.groupby(by=['Gender'])['Result'].mean()
ASD_gender_yes_count = autism_yes.groupby(by=['Gender'])['Gender'].count()
ASD_gender_no_count = autism_no.groupby(by=['Gender'])['Gender'].count()
ASD_gender_all_count = autism_data.groupby(by=['Gender'])['Gender'].count()

ASD_gender_df = pd.concat([ASD_gender_yes_percent,
                              ASD_gender_no_percent,
                              ASD_gender_yes_count,
                              ASD_gender_no_count,
                              ASD_gender_all_count,
                              ASD_gender_yes_score,
                              ASD_gender_no_score,
                              ASD_gender_all_score
                             ], axis=1, ignore_index = True, sort = True)

ASD_gender_df = ASD_gender_df.reset_index()
ASD_gender_df.columns = ['Gender','Percentage with Autism',
                            'Percentage without Autism',
                            'Number of People with Autism',
                            'Number of People without Autism',
                            'Total Number of People',
                            'Mean Autism Score for People with Autism',
                            'Mean Autism Score for People without Autism',
                            'Mean Autism Score for All People']
ASD_gender_df.head()
```
